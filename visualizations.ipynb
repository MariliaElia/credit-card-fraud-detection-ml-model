```python
# Imported Libraries
import numpy as np
import pandas as pd 

# Visual Libraries
from matplotlib import pyplot as plt
import seaborn as sns
plt.style.use('ggplot')

import warnings
warnings.filterwarnings('ignore')
```

```python
#Number of fraud and non fraud data
fraud = df[df['Class'] == 1]
nonFraud = df[df['Class'] == 0]

fraud_perc = len(fraud)/df.shape[0]
nonFraud_perc = len(nonFraud)/df.shape[0]

print("Total number of transactions is", df.shape[0],"of which", len(fraud),"are fraudulent and", len(nonFraud), "are not.")
print('Percentage of frauds :',fraud_perc*100,'%')
print('Percentage of non frauds :',nonFraud_perc*100,'%')
```

```python
#Graphs color theme
blue_color = '#4696CC' 
orange_color = '#F0884D'
```

```python
#Plot Transaction class against Frequency
labels = ['0 - Non Frauds','1 - Fraud']
classes = pd.value_counts(df['Class'], sort = True)
ax = classes.plot(kind = 'bar', rot=0, color=blue_color)
plt.title("Fraudulent vs. Non-Fraudulent Transactions", fontsize=14)
plt.xticks(range(2), labels, fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel("Class", fontsize=14)
plt.ylabel("Count", fontsize=14)

# Adding count on top of each bar
for i, count in enumerate(classes):
    ax.text(i, count + 50, str(count), ha='center', va='bottom', fontsize=12)

plt.show()
```

```python
# Create a displot for Time distribution by Class
g = sns.displot(data=df, x='Time', hue='Class', kind='kde', common_norm=False, palette={0: '#4696CC', 1: '#F0884D'})

# Set title and labels
plt.title("Time Distribution for Fraud and Non-Fraud Transactions", fontsize=14)
plt.xlabel("Time", fontsize=14)
plt.ylabel("Density", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Customize legend labels
g._legend.set_title('Class')
legend_labels = {0: 'Non-Fraud', 1: 'Fraud'}
for t, l in zip(g._legend.texts, legend_labels.values()):
    t.set_text(l)

g._legend.set_bbox_to_anchor((1.13, 0.5))
plt.show()
```

```python
# distribution of Time
time = df['Time'].values
sns.displot(time, color=blue_color)
plt.title("Time Distribution (Seconds)")
plt.yticks(fontsize=10)
plt.xticks(fontsize=9)
plt.xlabel("Time", fontsize=14)
plt.ylabel("Density", fontsize=14)
```

```python
# distribution of Amount
amount = [df['Amount'].values]
sns.distplot(amount, color=blue_color)
plt.title("Distribution of Amount", fontsize=14)
plt.xlabel("Amount", fontsize=14)
plt.ylabel("Density", fontsize=14)
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)
```

```python
# Set up the figure and axis
plt.figure(figsize=(8, 6))

# Create a boxplot for both fraudulent and non-fraudulent transactions
sns.boxplot(x='Class', y='Amount', data=df)

# Add labels and title
plt.xlabel('Class', fontsize=16)
plt.ylabel('Amount', fontsize=16)
plt.title('Boxplot of Amount by Transaction Class', fontsize=16)

# Add x-axis tick labels
plt.xticks([0, 1], ['Non-Fraud', 'Fraud'], fontsize=14)

plt.show()
```

```python
correlation_matrix = df.corr()

fig = plt.figure(figsize=(9,7))
sns.heatmap(correlation_matrix, cmap='RdBu_r')  # Ensure vmin=0 and vmax=1 for normalization
plt.show()
```

```python
# DIstribution plots of V1-V28 features split by fraudulent and non-fraudulent transactions
fig, axes = plt.subplots(6, 5, figsize=(12,9))
fig.supylabel('Density', fontsize=14)
fig.supxlabel('Features', fontsize=14)
fig.tight_layout()

target_0 = df.loc[df['Class'] == 0]
target_1 = df.loc[df['Class'] == 1]

for i, column in enumerate(df.columns[1:29], 1):
    plt.subplot(6,5,i)
    sns.distplot(target_0[[column]], hist=False, rug=True, color=blue_color)
    sns.distplot(target_1[[column]], hist=False, rug=True, color=orange_color)
    plt.xlabel(None)
    plt.ylabel(None)
    plt.title(column, fontsize=12)
    plt.xticks(fontsize=9)
    plt.yticks(fontsize=9)

plt.show()
```
